```python
# En una escuela se quiere tener un sistema para guardar la información de sus estudiantes para tener mejor organizado sus datos.

# Crear un diccionario que sirve para representar a una persona en este contexto, pensar en las características que se consideren más relevantes para identificar a una persona
# (su nombre, DNI, edad, etc).
estudiante = {
  'nombre': 'Juan',
  'apellido': 'Gonzalez',
  'dni': '48965412',
  'edad': 16
}

# Agregar al diccionario creado, un campo que sea otro diccionario y sirva para guardar el curso del estudiante y sus características (año, división, orientación, etc).
estudiante['curso'] = {
  'anio': '5to',
  'division': 'A',
  'turno': 'Mañana',
  'orientacion': 'Ciencias Sociales'
}

# Teniendo una lista de diccionarios de estudiantes, buscar en la lista la persona con mayor edad e imprimirla por pantalla.
estudiantes = [
  {
    'nombre': 'Juan',
    'apellido': 'Gonzalez',
    'dni': '48965412',
    'edad': 16,
    'curso': {
      'anio': '5to',
      'division': 'A',
      'turno': 'Mañana',
      'orientacion': 'Ciencias Sociales'
    }
  },
  {
    'nombre': 'Maria',
    'apellido': 'Perez',
    'dni': '48965413',
    'edad': 17,
    'curso': {
      'anio': '5to',
      'division': 'B',
      'turno': 'Tarde',
      'orientacion': 'Ciencias Naturales'
    }
  },
  {
    'nombre': 'Pedro',
    'apellido': 'Gomez',
    'dni': '48965414',
    'edad': 18,
    'curso': {
      'anio': '6to',
      'division': 'B',
      'turno': 'Noche',
      'orientacion': 'Economia'
    }
  }
]

mayor_edad = estudiantes[0]
for estudiante in estudiantes:
  if estudiante['edad'] > mayor_edad['edad']:
    mayor_edad = estudiante

print(mayor_edad)
```

```python
# En un vivero se guardan las plantas en una lista de diccionario con la siguiente información: especie, si necesita luz solar o no, y el precio.
plantas = [
  {
    'especie': 'Rosa',
    'necesita_sol': True,
    'precio': 200
  },
  {
    'especie': 'Margarita',
    'necesita_sol': False,
    'precio': 150
  },
  {
    'especie': 'Orquidea',
    'necesita_sol': True,
    'precio': 300
  }
]

# Ahora se necesita un sistema que guarde las plantas a medida que van llegando. Se pide hacer una función que reciba la lista de diccionarios de plantas,
# y los datos de la planta nueva y agregue esa planta a la lista de diccionarios.
def agregar_planta(plantas, especie, necesita_sol, precio):
  plantas.append({
    'especie': especie,
    'necesita_sol': necesita_sol,
    'precio': precio
  })

agregar_planta(plantas, 'Girasol', True, 100)

print(plantas)
```

```python
# Se representa un ticket de supermercado como una lista de diccionarios, donde cada diccionario tiene la siguiente información:
# Nombre del producto, precio por unidad y cantidad.
# Se pide hacer una función que reciba el ticket y devuelva el monto total a pagar.
def calcular_total(ticket):
  total = 0
  for producto in ticket:
    total += producto['precio'] * producto['cantidad']
  return total

ticket = [
  {
    'nombre': 'Arroz',
    'precio': 50,
    'cantidad': 2
  },
  {
    'nombre': 'Fideos',
    'precio': 40,
    'cantidad': 3
  },
  {
    'nombre': 'Pan',
    'precio': 30,
    'cantidad': 1
  }
]

print(f"Total a pagar: {calcular_total(ticket)}")
```

```python
# Sol tiene una lista de diccionarios donde guarda todas las películas que vió. La información que tiene para cada una es: el nombre de la serie, año en que salió,
# y la puntuación que le puso del 1 al 10. Hace mucho que quiere que Tomás empiece a ver las películas que ella considera que son las mejores que vio.
# Hacer una función que reciba el diccionario de las películas que vió Sol, y que devuelva una nueva lista de diccionarios donde sólo estén
# las películas que tienen puntaje mayor a 7.

pelis_vistas = [
  {
    'nombre': 'Pulp Fiction',
    'anio': 1994,
    'puntuacion': 9
  },
  {
    'nombre': 'El Señor de los Anillos',
    'anio': 2001,
    'puntuacion': 7
  },
  {
    'nombre': 'Interestelar',
    'anio': 2014,
    'puntuacion': 7
  },
  {
    'nombre': 'Titanic',
    'anio': 1997,
    'puntuacion': 6
  },
  {
    'nombre': 'Your Name',
    'anio': 2016,
    'puntuacion': 8
  },
  {
    'nombre': 'La La Land',
    'anio': 2016,
    'puntuacion': 9
  },
]

def mejores_peliculas(peliculas):
  mejores = []
  for pelicula in peliculas:
    if pelicula['puntuacion'] > 7:
      mejores.append(pelicula)
  return mejores

# Otra implementación usando list comprehension (https://www.geeksforgeeks.org/python-list-comprehension/)
def mejores_peliculas(peliculas):
  return [pelicula for pelicula in peliculas if pelicula['puntuacion'] > 7]

print(mejores_peliculas(pelis_vistas))
```

```python
# Un profesor guarda las notas del primer parcial de sus alumnos en una lista de diccionarios que guarda la siguiente información:
# Nombre, Apellido, Intento, Nota
# Donde ”intento” es la instancia que está rindiendo, 1 si es la primera vez que rinde el parcial, 2 si es el primer recuperatorio y 3 si es el segundo recuperatorio.
# Se pide hacer una función que, dado esta lista de diccionarios, devuelva el promedio de las notas en la primera oportunidad que rindieron los alumnos.

# ¿Cómo harían para generalizar la función y que el intento sea parametrizable? Es decir, que no solamente sirve para el intento 1,
# sino que también pueda servir para los demás.

notas = [
  {
    'nombre': 'Juan',
    'apellido': 'Gonzalez',
    'intento': 1,
    'nota': 7
  },
  {
    'nombre': 'Maria',
    'apellido': 'Perez',
    'intento': 1,
    'nota': 8
  },
  {
    'nombre': 'Pedro',
    'apellido': 'Gomez',
    'intento': 2,
    'nota': 6
  },
  {
    'nombre': 'Jose',
    'apellido': 'Rodriguez',
    'intento': 3,
    'nota': 9
  },
  {
    'nombre': 'Camila',
    'apellido': 'Martinez',
    'intento': 1,
    'nota': 8
  },
  {
    'nombre': 'Lucia',
    'apellido': 'Garcia',
    'intento': 2,
    'nota': 7
  }
]

def promedio_notas(notas, intento):
  notas_intento = []
  for nota in notas:
    if nota['intento'] == intento:
      notas_intento.append(nota['nota'])
  return sum(notas_intento) / len(notas_intento)

# Otra implementación usando list comprehension
def promedio_notas(notas, intento):
  notas_intento = [nota['nota'] for nota in notas if nota['intento'] == intento]
  return sum(notas_intento) / len(notas_intento)

print(promedio_notas(notas, 1))
```

```python
# En una fábrica, se hace un chequeo de calidad a los productos antes de cada entrega. El resultado del chequeo de la entrega se guarda en una lista de diccionarios,
# donde cada diccionario tiene la siguiente información de cada producto: Código del producto, Fecha de vencimiento, Si pasó el chequeo de calidad o no
# Se pide hacer una función que reciba esta lista de diccionarios y elimine todos los productos que no pasaron el chequeo de calidad.
# Devolver en una tupla el diccionario con los elementos eliminados y la cantidad de elementos que quedaron en el diccionario.
# Dado que la tupla es inmutable y nosotros no podemos ir agregando elementos a una tupla, ¿En qué momento deberíamos crear la tupla?
productos = [
  {
    'codigo': 10001,
    'fecha_vencimiento': '2022-12-01',
    'paso_chequeo': True
  },
  {
    'codigo': 10002,
    'fecha_vencimiento': '2022-12-01',
    'paso_chequeo': False
  },
  {
    'codigo': 10003,
    'fecha_vencimiento': '2022-12-01',
    'paso_chequeo': True
  },
  {
    'codigo': 10004,
    'fecha_vencimiento': '2022-12-01',
    'paso_chequeo': False
  }
]

def control_calidad(productos):
  pasaron_control = []
  no_pasaron_control = []
  for producto in productos:
    if producto['paso_chequeo'] == True:
      pasaron_control.append(producto)
    else:
      no_pasaron_control.append(producto)

  return (no_pasaron_control, len(pasaron_control))

print(control_calidad(productos))
```

```python
# Se quiere guardar la información de un grupo de maratonistas. Se necesita guardar su nombre, DNI, y todas las maratones que corrió,
# de la cual a su vez se quiere tener el nombre de cada una, el año, el puesto en que salió el maratonista, y el tiempo que tardó en terminarla.
# a. Crear el diccionario que represente esta situación.
# AYUDA: Queremos guardar muchos maratonistas, y a su vez, muchas maratones para cada maratonista, entonces
# ¿Qué tipo de dato debería ser el campo que guarda todas las maratones? ¿Y qué tipo de dato es la maratón en sí?
maratonistas = [
  {
    'nombre': 'Juan',
    'dni': 48965412,
    'maratones': [
      {
        'nombre': 'Maraton Buenos Aires',
        'anio': 2021,
        'puesto': 10,
        'tiempo_minutos': 120
      },
      {
        'nombre': 'Maraton Rosario',
        'anio': 2020,
        'puesto': 5,
        'tiempo_minutos': 110
      }
    ]
  },
  {
    'nombre': 'Maria',
    'dni': 48965413,
    'maratones': [
      {
        'nombre': 'Maraton Buenos Aires',
        'anio': 2021,
        'puesto': 15,
        'tiempo_minutos': 130
      },
      {
        'nombre': 'Maraton Cordoba',
        'anio': 2020,
        'puesto': 8,
        'tiempo_minutos': 115
      }
    ]
  },
  {
    'nombre': 'Pedro',
    'dni': 48965414,
    'maratones': [
      {
        'nombre': 'Maraton Cordoba',
        'anio': 2021,
        'puesto': 20,
        'tiempo_minutos': 140
      },
      {
        'nombre': 'Maraton Mar del Plata',
        'anio': 2020,
        'puesto': 12,
        'tiempo_minutos': 125
      }
    ]
  }
]

# b. Teniendo una lista de diccionarios de maratonistas, ordenarlos alfabéticamente.
maratonistas.sort(key=lambda maratonista: maratonista['nombre'])

# c. Ordenar las maratones de cada maratonista según el tiempo que tardó en completar cada una de forma ascendente.
for maratonista in maratonistas:
  maratonista['maratones'].sort(key=lambda maraton: maraton['tiempo_minutos'])

maratonistas
```
